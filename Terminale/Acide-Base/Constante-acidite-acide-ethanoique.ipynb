# Détermination de la constante d'acidité de l'acide éthanoïque

```python
import numpy as np
from scipy.optimize import curve_fit
from matplotlib import pyplot as plt
import seaborn as sns
```

## Variables du problème

```python
CA = 0.01  # mol/L
CB = 0.01  # mol/L
VA = [0.025, 0.020, 0.015, 0.010, 0.005]  # L
VB = [0.005, 0.010, 0.015, 0.020, 0.025]  # L
pH = [3.96, 4.33, 4.66, 4.98, 5.38]
```

## Détermination des valeurs de la variable x

```python
x = []
for i in range(len(VA)):
    x.append(np.log10(VB[i] /VA[i]))
```

## Tracé de l'évolution du pH

```python
sns.set()
plt.plot(x, pH, 'o')
plt.xlabel("x")
plt.ylabel("pH")
plt.show()
```

# Modélisation de l'évolution du pH

```python
# Fonction modèle
def modele(x, a, b):
    return a * x + b

# Détermination des paramètres optimaux
popt, pcov = curve_fit(modele, x, pH)
a_mod = popt[0]
b_mod = popt[1]

# Tracé
x_mod = np.linspace(min(x), max(x), 100)
y_mod = modele(x_mod, a_mod, b_mod)

sns.set()
plt.plot(x, pH, 'o')
plt.plot(x_mod, y_mod)
plt.xlabel("x")
plt.ylabel("pH")
plt.show()

print("a = {}, b = {}".format(a_mod, b_mod))
```
